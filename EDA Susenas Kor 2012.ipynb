## Data Cleansing

### Check Missing Value

Sebelum observasi lebih jauh, kami memeriksa kelengkapan data dari tiap variabel yang akan diambil. Jika terdapat banyak nilai kosong pada variabel tersebut, akan dilakukan imputasi untuk memudahkan proses analisis.

```python
def missing_value(dataset):

    total = dataset.isnull().sum().sort_values(ascending=False)
    percent = (dataset.isnull().sum()/dataset.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    
    return missing_data
```

```python
missing_value(df_raw)
```

```python
def separate_dtype(dataset,
                  CATEGORICAL_COLUMNS,
                  NUMERICAL_COLUMNS):
    categorical = dataset[CATEGORICAL_COLUMNS].copy()
    numerical = dataset[NUMERICAL_COLUMNS].copy()
    
    return categorical, numerical
```

6. __Mengkonversi nilai pada `kota_kabupaten`.__ Kode pada `kota_kabupaten` akan diubah menjadi nama-nama kota/kabupaten sesuai dengan data pemetaan daerah BPS tahun 2009. Karena banyaknya jumlah kode yang akan diubah, maka kami menambahkan satu dataframe yang nantinya akan berguna untuk proses *mapping*.

Setelah terbentuk list yang berisikan *key* dan juga *value*, list tersebut akan diubah menjadi dict dan menggunakan fungsi *mapping* untuk mengganti nilainya.

```python
def column_to_list(dataframe,var_name):
    ''' A function to take column values to list
    dataframe: str
        data
    var_name: str
        data variable name
    '''
    list_ = dataframe[var_name].tolist()
    set_ = set(list_)
    list_ = list(set_)
    
    return list_
```

```python
def making_dict(list_1,list_2):
    ''' A function to make dictionary from lists
    list_1: list
        list of string or integer
    list_2: list
        list of string
    '''
    dictionary = dict(zip(list_1,list_2))
    
    return dictionary
```

```python
def map_from_dict(var_name,dict_name):
    ''' A function to map series with dictionary
    var_name: str
        data variable name
    dict_name: str
        dictionary for mapping
    '''
    mapping = cat_data[var_name].map(dict_name)
    
    return mapping
```

### Data Level Rumah Tangga

Pertama, kami ingin melihat jenis kepemilikan rumah di tingkat pedesaan/perkotaan, jenis bangunan yang ditempati dan juga banyak ART yang menempati bangunan tempat tinggal. Karena tipe data yang digunakan adalah string, maka bentuk visualisasi yang digunakan adalah *bar chart*.

```python
plots = {1 : [111], 2: [121, 122], 3: [131, 132, 133], 4: [221, 222, 223, 224], 5: [231, 232, 233, 234, 235], 6: [231, 232, 233, 234, 235, 236], 7: [331, 332, 333, 334, 335, 336,337], 
         8: [331, 332, 333, 334, 335, 336, 337, 338], 9: [331, 332, 333, 334, 335, 336, 337, 338, 339]}

def countplot(x, y, data):
    
    rows = int(str(plots[len(y)][0])[0])
    columns = int(str(plots[len(y)][0])[1])
    plt.figure(figsize=(7*columns, 7*rows))
    
    for i, j in enumerate(y):
        plt.subplot(plots[len(y)][i])
        ax = sns.countplot(x=j, hue=x, data=data, palette='Dark2', alpha=0.8, linewidth=0.4, edgecolor='black')
        ax.set_title(j)
        
    return plt.show()
```

```python
countplot("Jenis Kepemilikan Rumah", ['Klasifikasi Desa/Kelurahan', 'Jenis Bangunan', 'Banyak ART'], df_fin)
```

Lalu untuk mengetahui pengaruh jumlah anak sekolah (<18 tahun) dan juga pengeluaran per kapita pada jenis kepemilikan rumah. Karena data yang digunakan adalah numerikal, maka visualisasi yang digunakan berbentuk *boxplot*.

```python
def boxplot(x, y, df):
    
    rows = int(str(plots[len(y)][0])[0])
    columns = int(str(plots[len(y)][0])[1])
    plt.figure(figsize=(20*columns, 10*rows))
    
    for i, j in enumerate(y):
        plt.subplot(plots[len(y)][i])
        ax = sns.boxplot(x=x, y=j, data=df[[x, j]], palette="Dark2", linewidth=1)
        ax.set_title(j)
        
    return plt.show()
```

```python
boxplot("Jenis Kepemilikan Rumah", ["Jumlah Anak Sekolah"], df_fin)
```

Berikutnya kami ingin melihat bagaimana persebaran jumlah anak sekolah dan juga pengeluaran per kapita yang akan divisualisasikan dengan *histogram chart*.

```python
def dist_plot(data, variabel):
    plot_umur = data[variabel]
    plot_umur.plot.hist(grid=False, bins=99, rwidth=1,color='blue',label=variabel)
    plt.title(variabel)
    plt.xlabel('Value ')
    plt.ylabel('Counts')
    plt.legend(loc="best")
    plt.tight_layout()
    plt.rcParams['figure.figsize'] = [10, 4]
    
    return plt.show()
```

```python
dist_plot(df_fin, 'Jumlah Anak Sekolah')
```

```python
dist_plot(df_fin, 'Pengeluaran Per Kapita (Log)')
```
